```python
# this is task2.py
import parser_data


def clean_data(data, num=None):
    cleaned = [i for i in data if i[0] > 10 and i[0] < 65]
    print(len(cleaned))
    file_name_clean = "walking_steps_clean.csv"
    if num is not None:
        file_name_clean = f"data/walking_steps_{num}_clean.csv" 
        with open(file_name_clean, "w") as f:
            f.write("time,X,Y,Z\n")
            for i in cleaned:
                f.write(str(i)[1:-1] + "\n")



def main():
    # Get data
    for i in [1]: # [1, 2] don't have second data set
        file_name = f"walking_steps_{i}.csv" 
        data = parser_data.get_data(file_name) 
        clean_data(data, i)
        newdata = parser_data.get_data(f"walking_steps_{i}_clean.csv")
        plot_data(data)
        plot_data(newdata)


if __name__== "__main__":
    main()

```
